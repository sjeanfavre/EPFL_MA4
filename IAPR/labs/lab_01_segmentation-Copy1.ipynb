### 2.1 Visualization

```python
# Load images
im_names = ['arena-shapes-01', 'arena-shapes-02', 'arena-shapes-03']
filenames = [os.path.join(data_path, name) + '.png' for name in im_names]
ic = skimage.io.imread_collection(filenames)
images = skimage.io.concatenate_images(ic)
print('Number of images: ', images.shape[0])
print('Image size: {}, {} '.format(images.shape[1], images.shape[2]))
print('Number of color channels: ', images.shape[-1])
```

```python
# Plot images
fig, axes = plt.subplots(1, 3, figsize=(12, 12))
for ax, im, nm in zip(axes.ravel(), images, im_names):
    ax.imshow(im)
    ax.axis('off')
    ax.set_title(nm)
plt.show()
```

This methoed doesn't work well since the uniform background creates a spike in the histogram. As a result, the eualized image is dominated by contrast. Let's try now the Contrast Limited Adaptive Histogram Equalization (CLAHE).

```python
clahe = cv.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))

images_HSV_clahe = images_HSV
images_HSV_clahe[0,:,:,2] = clahe.apply(images_HSV[0,:,:,2])
images_HSV_clahe[1,:,:,2] = clahe.apply(images_HSV[1,:,:,2])
images_HSV_clahe[2,:,:,2] = clahe.apply(images_HSV[2,:,:,2])
```

```python
# Plot histogram of value channel for the three images after histogram equalization using CLAHE
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, im, nm in zip(axes.ravel(), images_HSV_clahe, im_names):
    ax.hist(im[:,:,2].ravel(), bins=50)
    ax.set_title(nm)
plt.show()
```

```python
# Convert images back to RGB
images_RGB_clahe = np.array([cv.cvtColor(im, cv.COLOR_HSV2RGB) for im in images_HSV_clahe])
```
